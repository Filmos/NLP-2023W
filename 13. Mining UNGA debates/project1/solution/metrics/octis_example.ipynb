```python
from octis.dataset.dataset import Dataset
from octis.evaluation_metrics.coherence_metrics import Coherence
from octis.evaluation_metrics.diversity_metrics import TopicDiversity
from octis.models.LDA import LDA
```

```python
dataset = Dataset()
dataset.fetch_dataset("BBC_News", data_home="./octis_cache")
corpus = dataset.get_corpus()

corpus
```

```python
model = LDA()
output = model.train_model(dataset)

output
```

```python
coherence_metric = Coherence(corpus)
coherence_value = coherence_metric.score(output)

coherence_value
```

```python
diversity_metric = TopicDiversity()
diversity_value = diversity_metric.score(output)

diversity_value
```
